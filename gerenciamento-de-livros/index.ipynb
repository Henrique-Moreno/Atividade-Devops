```python
# Importando a biblioteca

import matplotlib.pyplot as plt
```

```python
# A classe Livro representa um livro com título, autor, gênero e quantidade disponível.
# Ela permite adicionar notas de avaliação, calcular a média das notas e determinar

class Livro:
    def __init__(self, titulo, autor, genero, quantidade):
        self.titulo = titulo  # Título do livro
        self.autor = autor    # Autor do livro
        self.genero = genero  # Gênero do livro
        self.quantidade = quantidade  # Quantidade disponível
        self.notas = []  # Lista para armazenar as notas do livro

    def adicionar_nota(self, nota):
        self.notas.append(nota)  # Adiciona uma nova nota à lista
        print(f'Nota {nota} adicionada ao livro "{self.titulo}".')

    def calcular_media(self):
        if not self.notas:  # Verifica se há notas
            return 0  # Retorna 0 se não houver notas
        return sum(self.notas) / len(self.notas)  # Calcula a média das notas

    def determinar_situacao(self):
        media = self.calcular_media()  # Calcula a média das notas
        if media >= 7:  # Se a média for maior ou igual a 7, o livro é bem avaliado
            return "Bem Avaliado"
        elif media >= 5:  # Se a média for maior ou igual a 5 e menor que 7, o livro é razoável
            return "Razoável"
        else:  # Se a média for menor que 5, o livro é mal avaliado
            return "Mal Avaliado"
```

```python
livros = []  # Lista que armazenará os objetos Livro
```

```python
def cadastrar_livro(titulo, autor, genero, quantidade):
    novo_livro = Livro(titulo, autor, genero, quantidade)  # Cria um novo objeto Livro
    livros.append(novo_livro)  # Adiciona o livro à lista
    print(f'Livro "{titulo}" cadastrado com sucesso!')  
```

```python
def listar_livros():
    if not livros:  # Verifica se a lista está vazia
        print("Nenhum livro cadastrado.")  
        return
    
    for livro in livros:  # Itera sobre cada livro na lista
        print(f'Título: {livro.titulo}, Autor: {livro.autor}, Gênero: {livro.genero}, Quantidade: {livro.quantidade}, Média: {livro.calcular_media():.2f}, Situação: {livro.determinar_situacao()}')
```

```python
def buscar_livro(titulo):
    for livro in livros:  # Itera sobre cada livro na lista
        if livro.titulo.lower() == titulo.lower():  # Compara o título (ignorando maiúsculas/minúsculas)
            print(f'Livro encontrado: Título: {livro.titulo}, Autor: {livro.autor}, Gênero: {livro.genero}, Quantidade: {livro.quantidade}, Média: {livro.calcular_media():.2f}, Situação: {livro.determinar_situacao()}')
            return
    
    print("Livro não encontrado.")
```

```python
# Cadastrar livros
cadastrar_livro("O Senhor dos Anéis", "J.R.R. Tolkien", "Fantasia", 5)
```

```python
cadastrar_livro("1984", "George Orwell", "Ficção", 3)
```

```python
cadastrar_livro("Pai Rico, Pai Pobre", "Robert Kiyosaki", "Finanças Pessoais", 14)
```

```python
# Listar todos os livros cadastrados
listar_livros()
```

```python
# Buscar um livro pelo título
buscar_livro("1984")
```

```python
def gerar_grafico_generos():
    generos = {}  # Dicionário para armazenar a quantidade de livros por gênero
    
    for livro in livros:
        if livro.genero in generos:
            generos[livro.genero] += livro.quantidade  # Soma a quantidade se o gênero já existir no dicionário
        else:
            generos[livro.genero] = livro.quantidade  # Adiciona o gênero ao dicionário
            
    plt.bar(generos.keys(), generos.values())  # Cria um gráfico de barras
    plt.xlabel('Gêneros')  # Rótulo do eixo X
    plt.ylabel('Quantidade de Livros')  # Rótulo do eixo Y
    plt.title('Quantidade de Livros por Gênero')  # Título do gráfico
    plt.show()  # Exibe o gráfico
```

```python
# Gerar gráfico de gêneros:
gerar_grafico_generos()
```

```python
def gerar_grafico_avaliacoes():
    titulos = [livro.titulo for livro in livros] # Cria uma lista com os títulos dos livros cadastrados
    medias = [livro.calcular_media() for livro in livros] # Cria uma lista com as médias das avaliações para cada livro
    
    plt.bar(titulos, medias)  
    plt.xlabel('Livros')  
    plt.ylabel('Média das Avaliações')  
    plt.title('Média das Avaliações dos Livros')  
    plt.xticks(rotation=45)  
    plt.show()  
```

```python
# Adicionar notas aos livros
livros[0].adicionar_nota(8) 
livros[0].adicionar_nota(9)   
livros[1].adicionar_nota(6)   
livros[1].adicionar_nota(4)   
```

```python
# Gerar gráfico de avaliações
gerar_grafico_avaliacoes()
```
